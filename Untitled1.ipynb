```python
from transformers import pipeline
```

```python
sentiment = pipeline("sentiment-analysis")
```

```python
print("Sentiment Analysis Program")
print("Type your sentences and see the sentiment. Type 'no' when you want to stop.\n")
results_summary = []

while True:
    try:
        text = input("Enter a sentence: ")

        # Stop if user types 'no'
        if text.strip().lower() == 'no':
            print("\nExiting program.\n")
            break

        # Get prediction
        result = sentiment(text)
        label = result[0]['label']
        score = result[0]['score']



        print(f"Sentiment: {label}, Confidence: {score:.4f}\n")

        # Save result for summary
        results_summary.append((text, label, score))

    except Exception as e:
        print(f"An error occurred: {e}\nPlease try again.\n")

# Display summary of all entered sentences
if results_summary != "":
    print("=== Summary of Entered Sentences ===")
    for idx, (sentence, label, score) in enumerate(results_summary, 1):
        print(f"{idx}. {sentence} → {label} (Confidence: {score:.4f})")

    # Optional: count of each sentiment
    pos_count = sum(1 for _, l, _ in results_summary if l == "POSITIVE")
    neg_count = sum(1 for _, l, _ in results_summary if l == "NEGATIVE")

    print(f"\nTotal Sentences: {len(results_summary)}")
    print(f"POSITIVE: {pos_count}, NEGATIVE: {neg_count}")
```
